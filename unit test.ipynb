```python
import pickle
import os
# Replace this with the path to your 'refs(unc).p' file
file_path = 'dataset/refer_seg/refcoco/refs(unc).p'

# Load the data
with open(file_path, 'rb') as file:
    refcoco_data = pickle.load(file)
```

```python
train=0
testA=0
testB=0
val=0
files=[]
val_files=[]
un=0
for r in refcoco_data:
    f = r['file_name'].split('_')[-2]

    if f not in files:
        
        files.append(f)
    
        if r['split']=='train':
            train+=1
        elif r['split']=='val':
            val_files.append(f)
            val+=1
        elif r['split']=='testA':
            testA+=1
        elif r['split']=='testB':
            testB+=1
        else:
            print(r['split'])
    else:
        print(f)
    
```
